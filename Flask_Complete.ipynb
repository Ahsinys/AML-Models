```python
%%writefile flask_aml_app/app.py

from flask import Flask, render_template, request, send_from_directory, redirect, url_for, session
from flask_cors import CORS
import os
import pandas as pd
import numpy as np

import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt

from sklearn.cluster import DBSCAN, KMeans
from sklearn.preprocessing import StandardScaler

from transformers import pipeline
from PIL import Image

import speech_recognition as sr
from gtts import gTTS

from mlxtend.frequent_patterns import apriori, association_rules

from markupsafe import Markup

from mlxtend.preprocessing import TransactionEncoder

# =========================
# APP INIT
# =========================
app = Flask(__name__)
CORS(app)

app.secret_key = "mysecretkey123"

app.config['UPLOAD_FOLDER'] = 'flask_aml_app/static/uploads'


# =========================
# LOAD MODELS ONCE
# =========================
gender_pipe = pipeline(
    "image-classification",
    model="rizvandwiki/gender-classification"
)

sentiment_pipe = pipeline("sentiment-analysis")

ner_pipe = pipeline(
    "token-classification",
    model="dslim/bert-base-NER",
    aggregation_strategy="simple"
)

textgen_pipe = pipeline("text-generation", model="distilgpt2")
qa_pipe = pipeline(
    "question-answering",
    model="distilbert-base-cased-distilled-squad"
)

translate_pipe = pipeline("translation_en_to_ur", model="Helsinki-NLP/opus-mt-en-ur")


# =========================
# ROUTES (BASIC STRUCTURE)
# =========================

@app.route('/')
def home():
    return render_template('index.html')


# -------- DBSCAN -------------------------------------------
@app.route('/dbscan', methods=['GET', 'POST'])
def dbscan():
    result_image = None

    if request.method == 'POST':
        file = request.files['file']

        if file:
            path = os.path.join(app.config['UPLOAD_FOLDER'], file.filename)
            file.save(path)

            # Read CSV
            df = pd.read_csv(path)

            # Keep only numeric values
            X = df.select_dtypes(include=[np.number]).values

            # Scale data
            scaler = StandardScaler()
            X_scaled = scaler.fit_transform(X)

            # DBSCAN model
            model = DBSCAN(eps=0.5, min_samples=5)
            labels = model.fit_predict(X_scaled)

            # Plot
            plt.figure(figsize=(6,4))
            plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=labels, cmap='rainbow')
            plt.title("DBSCAN Clustering")

            # Save plot
            output_path = os.path.join(app.config['UPLOAD_FOLDER'], 'dbscan_plot.png')
            plt.savefig(output_path)
            plt.close()

            result_image = 'dbscan_plot.png'

    return render_template('dbscan.html', result=result_image)






# -------- KMEANS ---------------------------------------------
@app.route('/kmeans', methods=['GET', 'POST'])
def kmeans():
    result_image = None

    if request.method == 'POST':
        file = request.files['file']
        k = int(request.form['k'])

        if file:
            path = os.path.join(app.config['UPLOAD_FOLDER'], file.filename)
            file.save(path)

            # Read CSV
            df = pd.read_csv(path)

            # Keep numeric columns only
            X = df.select_dtypes(include=[np.number]).values

            # Scale data
            scaler = StandardScaler()
            X_scaled = scaler.fit_transform(X)

            # KMeans model
            model = KMeans(n_clusters=k, random_state=42, n_init=10)
            labels = model.fit_predict(X_scaled)

            # Centroids
            centroids = model.cluster_centers_

            # Plot clusters
            plt.figure(figsize=(6,4))
            plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=labels, cmap='rainbow', s=50)

            # Plot centroids (BIG X MARKERS)
            plt.scatter(
                centroids[:, 0],
                centroids[:, 1],
                marker='X',
                s=200,
                c='black',
                label='Centroids'
            )

            plt.title("K-Means Clustering")
            plt.legend()

            # Save plot
            output_path = os.path.join(app.config['UPLOAD_FOLDER'], 'kmeans_plot.png')
            plt.savefig(output_path)
            plt.close()

            result_image = 'kmeans_plot.png'

    return render_template('kmeans.html', result=result_image)






# -------- CNN GENDER ----------------------------------------
@app.route('/cnn_gender', methods=['GET', 'POST'])
def cnn_gender():
    result = None
    confidence = None

    if request.method == 'POST':
        file = request.files['file']

        if file:
            path = os.path.join(app.config['UPLOAD_FOLDER'], file.filename)
            file.save(path)

            # Open image
            img = Image.open(path)

            # Predict
            output = gender_pipe(img)

            # Get best result
            top = output[0]

            result = top['label']
            confidence = round(top['score'] * 100, 2)

    return render_template(
        'cnn_gender.html',
        result=result,
        confidence=confidence
    )






# -------- SENTIMENT ----------------------------------
@app.route('/sentiment', methods=['GET', 'POST'])
def sentiment():
    text = None
    label = None
    score = None

    if request.method == 'POST':
        audio = request.files['audio']

        audio_path = os.path.join(app.config['UPLOAD_FOLDER'], "sentiment_audio.wav")
        audio.save(audio_path)

        # Speech to Text
        recognizer = sr.Recognizer()

        try:
            with sr.AudioFile(audio_path) as source:
                audio_data = recognizer.record(source)
                text = recognizer.recognize_google(audio_data)

            # Sentiment analysis
            result = sentiment_pipe(text)[0]

            label = result['label']
            score = round(result['score'] * 100, 2)

        except Exception as e:
            text = "Could not process audio"
            label = "ERROR"
            score = 0

    return render_template(
        'sentiment.html',
        text=text,
        label=label,
        score=score
    )






# -------- QA ----------------------------------------------
@app.route('/qa', methods=['GET', 'POST'])
def qa():
    answer = None
    context_text = None
    question_text = None
    audio_file = None

    if request.method == 'POST':

        # TEXT INPUT (context)
        context_text = request.form['context']

        # AUDIO INPUT (question)
        question_audio = request.files['question_audio']

        q_path = os.path.join(app.config['UPLOAD_FOLDER'], "question.wav")
        question_audio.save(q_path)

        recognizer = sr.Recognizer()

        # Convert question audio → text
        with sr.AudioFile(q_path) as source:
            audio_data = recognizer.record(source)
            question_text = recognizer.recognize_google(audio_data)

        # QA MODEL
        result = qa_pipe({
            'question': question_text,
            'context': context_text
        })

        answer = result['answer']

        # TEXT → SPEECH (OUTPUT VOICE)
        tts = gTTS(answer, lang='en')

        audio_path = os.path.join(app.config['UPLOAD_FOLDER'], "qa_answer.mp3")
        tts.save(audio_path)

        audio_file = "qa_answer.mp3"

    return render_template(
        'qa.html',
        answer=answer,
        context=context_text,
        question=question_text,
        audio_file=audio_file
    )






# -------- TEXT GENERATION -------------------------------------
@app.route('/textgen', methods=['GET', 'POST'])
def textgen():
    generated_text = None
    prompt = None
    length = None

    if request.method == 'POST':
        prompt = request.form['prompt']
        length = int(request.form['length'])

        result = textgen_pipe(
            prompt,
            max_length=length,
            num_return_sequences=1
        )

        generated_text = result[0]['generated_text']

    return render_template(
        'textgen.html',
        generated_text=generated_text,
        prompt=prompt,
        length=length
    )






# -------- TRANSLATION --------------------------------------
@app.route('/translation', methods=['GET', 'POST'])
def translation():
    translated_text = None
    input_text = None

    if request.method == 'POST':
        input_text = request.form['text']

        result = translate_pipe(input_text)
        translated_text = result[0]['translation_text']

    return render_template(
        'translation.html',
        input_text=input_text,
        translated_text=translated_text
    )





# -------- NER ----------------------------------------------
@app.route('/ner', methods=['GET', 'POST'])
def ner():
    highlighted_text = None
    entities = []

    if request.method == 'POST':
        text = request.form['text']

        ner_results = ner_pipe(text)

        entities = ner_results

        # Sort entities by position (IMPORTANT)
        ner_results = sorted(ner_results, key=lambda x: x['start'])

        highlighted_text = ""
        last_idx = 0

        for ent in ner_results:
            start = ent['start']
            end = ent['end']
            label = ent.get('entity_group', 'MISC')

            # add normal text before entity
            highlighted_text += text[last_idx:start]

            # add highlighted entity
            highlighted_text += f'<span class="entity-{label}">{text[start:end]}</span>'

            last_idx = end

        # add remaining text
        highlighted_text += text[last_idx:]

        highlighted_text = Markup(highlighted_text)

    return render_template(
        'ner.html',
        highlighted_text=highlighted_text,
        entities=entities
    )






# -------- APRIORI ---------------------------------------------
@app.route('/apriori', methods=['GET', 'POST'])
def apriori_page():
    rules_data = []

    try:
        if request.method == 'POST':
            file = request.files.get('file')

            if not file or file.filename == '':
                return render_template('apriori.html', rules=[])

            path = os.path.join(app.config['UPLOAD_FOLDER'], "apriori.csv")
            file.save(path)

            df = pd.read_csv(path, header=None).dropna()

            transactions = [
                [x.strip() for x in str(row).split(',') if x.strip()]
                for row in df.iloc[:, 0]
            ]

            if len(transactions) == 0:
                return render_template('apriori.html', rules=[])

            te = TransactionEncoder()
            df_encoded = pd.DataFrame(
                te.fit(transactions).transform(transactions),
                columns=te.columns_
            )

            freq_items = apriori(df_encoded, min_support=0.001, use_colnames=True)

            if freq_items.empty:
                return "NO FREQUENT ITEMS FOUND"

            rules = association_rules(freq_items, metric='lift', min_threshold=1.0)

            rules_data = rules[
                ['antecedents', 'consequents', 'support', 'confidence', 'lift']
            ].to_dict('records')

            return render_template('apriori.html', rules=rules_data)

        return render_template('apriori.html', rules=[])

    except Exception as e:
        import traceback
        traceback.print_exc()
        return f"ERROR: {str(e)}"



# =========================
# RUN APP
# =========================
if __name__ == "__main__":
    app.run(debug=True)
```
